## Chunking

```python
import nltk
from nltk import word_tokenize, pos_tag, RegexpParser
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python

# Sample text
text = "The quick brown fox jumps over the lazy dog"

# Tokenize the text and perform POS tagging
words = word_tokenize(text)
pos_tags = pos_tag(words)

# Define a chunk grammar using regular expressions
chunk_grammar = r"""
  NP: {<DT>?<JJ>*<NN>}   # chunk determiner, adjectives, and noun
  PP: {<IN><NP>}         # chunk preposition followed by noun phrase
  VP: {<VB.*><NP|PP>*}   # chunk verb and its arguments
"""

# Create a chunk parser using the chunk grammar
chunk_parser = RegexpParser(chunk_grammar)

# Perform chunking
chunked_text = chunk_parser.parse(pos_tags)

# Print the chunked text
print(chunked_text)
```

- (NP The/DT quick/JJ brown/NN): This chunk represents a noun phrase (NP) consisting of the words "The quick brown."

- (NP fox/NN): This chunk represents another noun phrase (NP) containing the word "fox."

- (VP jumps/VBZ (PP over/IN (NP the/DT lazy/JJ dog/NN)): This chunk represents a verb phrase (VP) with the word "jumps," followed by a prepositional phrase (PP) "over the lazy dog," where the prepositional phrase itself contains a nested noun phrase (NP) "the lazy dog."

## TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample documents
documents = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?",
]

# Create a TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit the vectorizer to the documents and transform the documents into TF-IDF vectors
tfidf_matrix = vectorizer.fit_transform(documents)

# Get the feature names (terms)
feature_names = vectorizer.get_feature_names_out()

# Print the TF-IDF matrix
print("TF-IDF Matrix:")
print(tfidf_matrix.toarray())

# Print the feature names
print("\nFeature Names:")
print(feature_names)
```

```python
doc_a = 'this document is first document'
doc_b = 'this document is the second document'

bag_of_words_a = doc_a.split(' ')
bag_of_words_b = doc_b.split(' ')

unique_words_set = set(bag_of_words_a).union(set(bag_of_words_b))
print(unique_words_set)

# Now create a dictionary of words and their occurence for each document in the corpus (collection of documents).

dict_a = dict.fromkeys(unique_words_set, 0)
# print(dict_a) # {'this': 0, 'document': 0, 'second': 0, 'is': 0, 'the': 0}

for word in bag_of_words_a:
    dict_a[word] += 1

print(dict_a)
# {'this': 1, 'document': 2, 'second': 1, 'is': 1, 'the': 1}

# similarly

dict_b = dict.fromkeys(unique_words_set, 0)

for word in bag_of_words_b:
    dict_b[word] += 1

print(dict_b)
# D1 'this document is first document'
# D2 'this document is the second document'
```

There are some potential problems that might arise with the Bag of Words model when it is used on large corpora. Since the feature vectors are based on absolute term frequencies, there might be some terms that occur frequently across all documents and these may tend to overshadow other terms in the feature set. Especially words that don’t occur as frequently, but might be more interesting and effective as features to identify specific categories. This is where TF-IDF comes into the picture.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Given documents
doc_a = 'this document is first document'
doc_b = 'this document is the second document'

# Create a list of documents
documents = [doc_a, doc_b]

# Create a TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit the vectorizer to the documents and transform the documents into TF-IDF vectors
tfidf_matrix = vectorizer.fit_transform(documents)

# Get the feature names (terms)
feature_names = vectorizer.get_feature_names_out()

# Print the TF-IDF matrix
print("TF-IDF Matrix:")
print(tfidf_matrix.toarray())

# Print the feature names
print("\nFeature Names:")
print(feature_names)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Given documents
doc_a = 'this document is first document'
doc_b = 'this document is the second document'

# Create a list of documents
documents = [doc_a, doc_b]

# Create a TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit the vectorizer to the documents and transform the documents into TF-IDF vectors
tfidf_matrix = vectorizer.fit_transform(documents)

# Get the feature names (terms)
feature_names = vectorizer.get_feature_names_out()

# Create a dictionary to store the TF-IDF scores for each term in the documents
tfidf_scores = {}

# Iterate through each document and its corresponding TF-IDF vector
for i, doc in enumerate(documents):
    feature_index = tfidf_matrix[i, :].nonzero()[1]
    tfidf_scores_for_doc = {}
    for idx in feature_index:
        term = feature_names[idx]
        tfidf_score = tfidf_matrix[i, idx]
        tfidf_scores_for_doc[term] = tfidf_score
    tfidf_scores[f"doc_{i+1}"] = tfidf_scores_for_doc

# Print the TF-IDF scores for each document
for doc, scores in tfidf_scores.items():
    print(f"{doc}: {scores}")
```

TF-IDF (Term Frequency-Inverse Document Frequency) has several advantages over the simple bag of words approach:
Term Weighting: TF-IDF considers the importance of terms in a document by assigning weights based on their frequency and rarity across the entire corpus. This allows TF-IDF to capture the significance of terms in a document, whereas the bag of words approach treats all terms equally.
Normalization: TF-IDF normalizes the term frequencies by taking into account the inverse document frequency, which helps in mitigating the impact of common terms that appear frequently across documents. This normalization is not present in the bag of words approach.
Focus on Key Terms: TF-IDF emphasizes terms that are frequent in a document but rare in the corpus, effectively highlighting key terms that are more discriminative and informative. In contrast, the bag of words approach may give equal importance to common terms that do not carry much discriminatory power.
Reduced Dimensionality: TF-IDF tends to produce sparse vectors, especially for large vocabularies, as it only considers the non-zero TF-IDF scores. This can lead to reduced dimensionality compared to the bag of words representation, which can be beneficial for computational efficiency and memory usage.
Improved Information Retrieval: TF-IDF is commonly used in information retrieval systems to rank the relevance of documents to a query. By considering the importance of terms, TF-IDF can lead to more accurate retrieval results compared to the bag of words approach.
Overall, the advantages of TF-IDF over the bag of words approach include improved term weighting, normalization, focus on key terms, potential dimensionality reduction, and enhanced performance in information retrieval tasks.

### Lower casing

```python
def convert_to_lowercase(text):
    # Convert text to lowercase
    lowercase_text = text.lower()
    return lowercase_text

data['Comment'] = data['Comment'].apply(convert_to_lowercase)
```

### Punctution removal

```python
import string

def remove_punctuation(text):
    if isinstance(text, float):  # Check if the input is a float
        return text  # Return the input as it is
    else:
        ans = ""
        for char in text:
            if char not in string.punctuation:
                ans += char
        return ans  # Return the text with punctuation removed

data['Comment'] = data['Comment'].apply(remove_punctuation)
```

### Stopwords removal

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
```

```python
import nltk
from nltk.corpus import stopwords

def remove_stopwords(text):
    x = []
    for i in text.split():
        if i not in stopwords.words('english'):
            x.append(i)
    return x

data['Comment'] = data['Comment'].apply(remove_stopwords)
```
